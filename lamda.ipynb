```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'nama_kabupaten_kota': ['Bandung', 'Bogor', 'Bekasi', 'Depok', 'Sukabumi'],
    'jumlah_produksi_sampah': [1200, 1500, 1300, 900, 700]
}
df = pd.DataFrame(data)

colors = plt.cm.viridis(np.linspace(0, 1, len(df)))

plt.figure(figsize=(20, 12))
bars = plt.barh(df['nama_kabupaten_kota'], df['jumlah_produksi_sampah'], color=colors, edgecolor='black')

for bar in bars:
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f"{int(bar.get_width())} ton", va='center', fontsize=14, color='black')

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlabel('Jumlah Produksi Sampah (ton)', fontsize=16, fontweight='bold')
plt.ylabel('Kota/Kabupaten', fontsize=16, fontweight='bold')
plt.title('Produksi Sampah Kota/Kabupaten di Jawa Barat', fontsize=20, fontweight='bold')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# 2. Jumlahkan total sampah per kab/kota setiap tahun

pd.set_option('display.max_rows', None)

df = pd.read_csv('sampahjawabarat.csv')
df

sampah_kota_tahunan = df.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_produksi_sampah'].sum().reset_index()

print("Total Sampah per Kabupaten/Kota di Jawa Barat Setiap Tahun")
sampah_kota_tahunan
```

```python
total_sampah_tahunan = df.groupby('tahun')['jumlah_produksi_sampah'].sum()
colors = plt.colormaps['tab20'](np.linspace(0, 1, len(total_sampah_tahunan)))
total_sampah_tahunan.plot(kind='barh', color='blue')
plt.title('Total Sampah per Kab/Kota Setiap Tahun di Jawa Barat')
plt.xlabel('Tahun')
plt.ylabel('Total Produksi Sampah (ton)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Buatkan rata-rata menggunakan lambda per kabupaten kota

rata_sampah = df.groupby(['nama_kabupaten_kota'])['jumlah_produksi_sampah'].sum().reset_index()

rata_sampah['Rata-rata'] = df.apply(lambda row: df[df['nama_kabupaten_kota'] == row['nama_kabupaten_kota']]['jumlah_produksi_sampah'].mean(), axis=1)

print("hasi rata' sampah yang ada di kota jawabarat")
rata_sampah
```

```python
# 4. Buat kolom baru 'Kategori', buat if jika jumlah sampah lebih dari 400 ton maka sampah harus segera ditanggulangi 
# sedangkan jika tidak, sampah masih dalam kategori aman

pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', 1000)

def kategori_sampah(jumlah):
    return "Segera tanggulangi" if jumlah > 400 else "Aman"

df['Kategori'] = df['jumlah_produksi_sampah'].apply(kategori_sampah)
print("Kategori Sampah :")
df[['nama_kabupaten_kota', 'tahun', 'jumlah_produksi_sampah', 'Kategori']]
```

```python
# 5. Hitung bayaran sampah dimana 1 ton harganya adalah 123.000

df['bayaran_sampah'] = df['jumlah_produksi_sampah'] * 123.000
print("Bayaran Sampah :")
df[['nama_kabupaten_kota', 'tahun', 'jumlah_produksi_sampah', 'bayaran_sampah']]
```

```python
# 6. Hitung kategori sampah dan tambahkan kolom pajak jika sampah harus segera ditanggulangi

def kategori_sampah(jumlah):
    return "Segera Ditanggulangi" if jumlah > 400 else "Aman"

df['Kategori'] = df['jumlah_produksi_sampah'].apply(kategori_sampah)

df['Pajak'] = df.apply(lambda row: row['bayaran_sampah'] * 0.05 if row['Kategori'] == "Segera Ditanggulangi" else 0, axis=1)

print("Pajak sampah yang harus segera ditanggulangi :")
df[['nama_kabupaten_kota', 'tahun', 'jumlah_produksi_sampah', 'Kategori', 'bayaran_sampah', 'Pajak']]
```
